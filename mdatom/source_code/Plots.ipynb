```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename_normal_t1 = "out_normal_T120.27.csv"
filename_metro_t1 = "out_metro_T120.27.csv"



df_normal = pd.read_csv(filename_normal_t1, header=None)
df_metro = pd.read_csv(filename_metro_t1, header=None)

start_at = 0

plt.figure(1)
plt.ylabel('Potential energy')
plt.xlabel('Step')
plt.plot(df_normal.values[start_at:,0], df_normal.values[start_at:,1], df_metro.values[start_at:,0], df_metro.values[start_at:,1])

plt.figure(2)
plt.ylabel('Delta Potential energy')
plt.xlabel('Step')
plt.plot(df_normal.values[30:,0], df_normal.values[30:,2], df_metro.values[30:,0], df_metro.values[30:,2])


plt.figure(3)
plt.ylabel('R')
plt.xlabel('step')
plt.plot(df_metro.values[start_at:,0], df_metro.values[start_at:,3])
```

```python
filename_normal_t1_dis = "out_normal_T120.27_dis.csv"
filename_metro_t1_dis = "out_metro_T120.27_dis.csv"


plt.figure(3)
plt.xlabel('r')
plt.ylabel('g(r)')


df_normal_dis = pd.read_csv(filename_normal_t1_dis, header=None)
df_metro_dis = pd.read_csv(filename_metro_t1_dis, header=None)

plt.plot(df_normal_dis.values[30:,0], df_normal_dis.values[30:,1], df_metro_dis.values[30:, 0],df_metro_dis.values[30:,1])

```
