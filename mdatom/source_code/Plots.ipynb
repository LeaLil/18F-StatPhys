```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
filename_normal_t1 = "out_normal_T120.27.csv"
filename_metro_N = "out_metro_T120.27_N.csv"
filename_metro_1 = "out_metro_T120.27_1.csv"



df_normal = pd.read_csv(filename_normal_t1, header=None)
df_metro = pd.read_csv(filename_metro_N, header=None)
df_metro_1 = pd.read_csv(filename_metro_1, header=None)

start_at = 5

#N Particle
plt.figure(1)
fig, ax1 = plt.subplots()
t = df_normal.values[start_at:,0]

lf_e = ax1.plot(t, df_normal.values[start_at:, 1], color='blue', linestyle='dashed', label='Leap-Frog, E')
mc_e = ax1.plot(t, df_metro.values[start_at:, 1], color='blue', label='MC, E')
ax1.set_xlabel('Step')
ax1.set_ylabel('Potential Energy [kJ/mol]')
ax1.tick_params('y', colors='b')
ax2 = ax1.twinx()
lf_de = ax2.plot(t, df_normal.values[start_at:, 2], color='olivedrab', linestyle='dashed', label='Leap-Frog, E')
mc_de = ax2.plot(t, df_metro.values[start_at:, 2], color='olivedrab', label='MC,  E')
ax2.set_ylabel('Delta Potential Energy [kJ/mol]')
ax2.tick_params('y', colors='olivedrab')
ax1.legend(loc='upper right')
ax2.legend(loc=(0.35,0.825))
plt.show()


#1 Particle
plt.figure(2)
fig, ax1 = plt.subplots()
t = df_normal.values[start_at:,0]

ax1.plot(t, df_normal.values[start_at:, 1], color='blue', linestyle='dashed', label='Leap-Frog, E')
ax1.plot(t, df_metro_1.values[start_at:, 1], color='blue', label='MC, E')
ax1.set_xlabel('Step')
ax1.set_ylabel('Potential Energy [kJ/mol]')
ax1.tick_params('y', colors='b')
ax2 = ax1.twinx()
ax2.plot(t, df_normal.values[start_at:, 2], color='olivedrab', linestyle='dashed', label='Leap-Frog, E')
ax2.plot(t, df_metro_1.values[start_at:, 2], color='olivedrab', label='MC,  E')
ax2.set_ylabel('Delta Potential Energy [kJ/mol]')
ax2.tick_params('y', colors='olivedrab')
ax1.legend(loc='upper right')
ax2.legend(loc=(0.35,0.825))

plt.show()

```
